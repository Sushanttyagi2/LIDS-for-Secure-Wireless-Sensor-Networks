```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 1. Load the dataset


```
df = pd.read_csv('BotNeTIoT-L01_label_NoDuplicates.csv')
```

# 2. Basic Dataset Health Check

```
print("Dataset Shape:", df.shape)
print("\nMissing Values:\n", df.isnull().sum().sum())
print("\nClass Distribution (0=Attack, 1=Normal):\n", df['label'].value_counts())
```

# 3. Visualise Class Balance

```
plt.figure(figsize=(6, 4))
sns.countplot(x='label', data=df, palette='viridis')
plt.title('Class Distribution: Normal (1) vs Attack (0)')
plt.show()
```

 # 4. Correlation Heatmap (Lightweight Feature Selection)
  We only look at features highly correlated with the 'label'

```
plt.figure(figsize=(12, 10))
# Select only numeric columns for correlation
corr_matrix = df.select_dtypes(include=['number']).corr()
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()
```

# 5. Identifying Top Features
 This helps you discard useless features to keep the IDS "lightweight"

```
top_features = corr_matrix['label'].sort_values(ascending=False).head(10)
print("\nTop 10 Features Correlated with Label:\n", top_features)
```
