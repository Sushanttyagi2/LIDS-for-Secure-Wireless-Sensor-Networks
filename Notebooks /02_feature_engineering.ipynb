```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest, f_classif
import joblib
```

# 1. Load your processed data

```python
df = pd.read_csv('BotNeTIoT-L01_label_NoDuplicates.csv')
```


# 2. Separate Features (X) and Target (y)


```python
X = df.drop(columns=['label'])
y = df['label']
```

# 3. Feature Selection (Crucial for "Lightweight" IDS)
 We select the top 10 most impactful features out of 23 to reduce computation


```python
selector = SelectKBest(score_func=f_classif, k=10)
X_selected = selector.fit_transform(X, y)
```

# Get the names of the selected features


```python
selected_features = X.columns[selector.get_support()]
print("Selected Lightweight Features:", list(selected_features))
```

# 4. Scaling (Normalizing data between 0 and 1)
 This is vital for WSNs so that large numbers don't overwhelm the model logic


```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_selected)
```

# 5. Split into Training and Testing sets


```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

# 6. Save the Scaler and Feature List (Very important for the /models folder!)
 You will need these to process real-time data later


```python
joblib.dump(scaler, 'models/scaler.pkl')
joblib.dump(selected_features, 'models/selected_features.pkl')

print(f"\nFeature engineering complete.")
print(f"Training shape: {X_train.shape}")
```
